## 2. The question: decision, action, cost of a wrong call

## Research Question

### Search Question

Which content pages should be refreshed first to improve search performance and user engagement?

### Unit of Analysis

Each row in the dataset represents one content page.

### Decision

The decision is which pages should be prioritized for refreshing.

### Who Takes the Action?

The content or SEO team reviews the highest-priority pages and updates their content, titles, metadata, keywords, or structure.

### Cost of a Wrong Recommendation

If a well-performing page is unnecessarily selected, time and editorial resources are wasted.

## 3. Quick look at the data (2-3 real numbers)


The starter dataset contains thousands of content pages collected from multiple clients.

These pages include information about search performance, engagement, freshness, traffic, and content characteristics.

The statistics below show that the dataset is large enough and contains sufficient variation to support a content opportunity scoring project.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/abdulmoeed1090/Fly-rank-starter-assignment/main/data/raw/content_refresh_anonymized.csv"

df = pd.read_csv(url)
df.head()

print("Rows:", len(df))
print("Columns:", len(df.columns))
print("Clients:", df["client_id"].nunique())

print("\nAverage Search Volume:")
print(round(df["search_volume"].mean(), 2))

print("\nAverage CTR:")
print(round(df["ctr"].mean(), 4))

print("\nAverage Content Age (in days):")
print(round(df["content_age_days"].mean(), 2))




plt.figure(figsize=(8,5))
df["content_age_days"].hist(bins=30)

plt.title("Distribution of Content Age")
plt.xlabel("Content Age (Days)")
plt.ylabel("Number of Pages")

plt.show()
```

However, it cannot prove that refreshing a page will increase traffic or rankings. Search performance is also affected by external factors such as algorithm updates, competition, seasonality, and user behavior.

The model should therefore be used as a ranking tool to help editors prioritize their work instead of replacing human judgment.
